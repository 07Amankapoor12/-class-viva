# Question 5
## Create a music of your choice and export the audio in .wav format

```python
import thinkdsp as t
```

```python
sa=t.SinSignal(freq=240,amp=5,offset=0)
sa_wave=sa.make_wave(duration=1,start=0,framerate=42000)
re=t.SinSignal(freq=270,amp=5,offset=0)
re_wave=re.make_wave(duration=1,start=1,framerate=42000)
ga=t.SinSignal(freq=300,amp=5,offset=0)
ga_wave=ga.make_wave(duration=1,start=2,framerate=42000)
ma=t.SinSignal(freq=320,amp=5,offset=0)
ma_wave=ma.make_wave(duration=1,start=3,framerate=42000)
pa=t.SinSignal(freq=350,amp=5,offset=0)
pa_wave=pa.make_wave(duration=1,start=4,framerate=42000)
dha=t.SinSignal(freq=370,amp=5,offset=0)
dha_wave=dha.make_wave(duration=1,start=5,framerate=42000)
ne=t.SinSignal(freq=410,amp=5,offset=0)
ne_wave=ne.make_wave(duration=1,start=6,framerate=42000)
sa2=t.SinSignal(freq=450,amp=5,offset=0)
sa_2_wave=sa2.make_wave(duration=1,start=7,framerate=42000)

song=sa_wave+re_wave+ga_wave+ma_wave+pa_wave+dha_wave+ne_wave+sa_2_wave
```

```python
song.write('Audio.wav')
```

# Question 1
## Create a mixture of 3 sinusoids of different linear frequencies which will be you date of birth, hen find Fourier transform to identify the frequencies.

```python
import numpy as np
import matplotlib.pyplot as plt
a=[5*np.sin(2*7*np.pi*t)for t in np.linspace(0,5,1000)]
b=[5*np.sin(2*12*np.pi*t)for t in np.linspace(0,5,1000)]
c=[5*np.sin(2*99*np.pi*t)for t in np.linspace(0,5,1000)]
mix=[(i+j+k) for i,j,k in zip(a,b,c)]
plt.plot(mix)
sig=np.fft.fft(mix)
```
